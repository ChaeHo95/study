```python
import requests
from  bs4 import BeautifulSoup 
import re

def cleanText(strs):
    text = re.sub('[\n\t\s]','',strs)
    return text

souce = requests.get('http://www.yjjob.or.kr/p/?j=23')
html = souce.text
soup = BeautifulSoup(html,'html.parser') 

items = soup.select('.r-eview')

itemlist = []
dic = {}

for item in items:
    ilist = []
    load = item.select_one('.font11').text
    title = item.select_one('a').attrs['title']
    for itemSun in item:
        if cleanText(itemSun.text) == cleanText(load)+cleanText(title):
            ilist.append(cleanText(load)) 
            ilist.append(cleanText(title )) 
        else:
            ilist.append(cleanText(itemSun.text)) 
    itemlist.append(ilist)

for item in itemlist:
    for i in range(len(item)):
        if i == 1:
            dic['과정구분']=item[i]
        elif i == 2:
            dic['과정명']=item[i]
        elif item[i] != '':
            dicItem = item[i].split(':',1)
            dic[dicItem[0]] = dicItem[1]
    for st in dic:
        print(st + " : "+ dic[st])
    print('-----------------------------------')
```
