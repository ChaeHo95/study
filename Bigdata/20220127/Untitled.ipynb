```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import konlpy
from konlpy.tag import Okt
from collections import Counter
from PIL import Image
from wordcloud import WordCloud

def image_make(content_img,content,num):
    img = Image.open('SouthKorea.jpg')
    img_array = np.array(img)
    wc = WordCloud(font_path='malgun', width=1200, height=1200, scale=4.0, max_font_size=500, mask=img_array, background_color = "white")
    gen = wc.generate_from_frequencies(content_img)
    plt.figure()
    plt.imshow(gen)
    plt.axis('off')                          # 축 설정
    plt.savefig(str(categorys[num])+content+'.png')       # 워드 클라우드 저장

def urlmake(links):
    source = links
    driver.get(source)
    req = driver.page_source
    soup = BeautifulSoup(req, "html.parser") 
    return soup

def isNone(word):
    if word == None and word.text == '':
        return '없음'
    else:
        return re.sub('[\t\n]','',word.text)
    
def text_cleaning(text):
    # 한글과 띄어쓰기를 제외한 모든 글자 패턴을 지정하여 hangul로 정의
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') 
    result = hangul.sub('', text) # 한글과 띄어쓰기를 제외한 모든 글자 패턴을 '' 빈 문자로 치환
    return result

def word_count(text):
    text_nouns = Okt().nouns(text)
    count = Counter(text_nouns)
    remove_text = Counter({x : count[x] for x in count if len(x) > 1})
    return remove_text
    
    
from selenium import webdriver
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

excutable_path = "chromedriver.exe"
driver = webdriver.Chrome(executable_path=excutable_path)

base_url = 'https://news.naver.com/'

search_lists = urlmake(base_url).select('.Nlist_item > a')

columns = [ "category","title", "content_text"]
df = pd.DataFrame(columns=columns)
categorys = []

for i in range(1,6):
    news_soup = urlmake(search_lists[i].attrs['href'])
    news_category = isNone(news_soup.select_one('.snb>h2'))
    news_contents = news_soup.select('#section_body> ul> li> dl> dt:nth-child(2)> a')
    categorys.append(news_category.replace('/',' '))
    for news in news_contents:
        news_title = isNone(news)
        new_url = news.attrs['href']
        news_content = isNone(urlmake(base_url + news.attrs['href']).select_one('#articleBodyContents'))
        row = [news_category,news_title, news_content]
        series = pd.Series(row, index=df.columns)
        df = df.append(series, ignore_index=True)
        
driver.close()
```

```python
df['title'] = df['title'].apply(text_cleaning)
df['content_text'] = df['content_text'].apply(text_cleaning)

title_corpus = []
content_corpus = []

# _ 필요없는 값을 버리겠다.
for _, category in df.groupby('category'):
    title_corpus .append(''.join(category['title'].tolist()))
    content_corpus.append(''.join(category['content_text'].tolist()))

categorys.sort()
print(categorys)
print(f'{categorys[0]}분야의 뉴스에 제목에 사용된 단어')
image_make(word_count(title_corpus[0]),'_제목',0)
```

```python
print(f'{categorys[0]}분야의 뉴스에 본몬에 사용된 단어')
image_make(word_count(content_corpus[0]),'_본몬',0)
```
