```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def urlmake(links):
    source = links
    driver.get(source)
    req = driver.page_source
    soup = BeautifulSoup(req, "html.parser") 
    return soup

def isNone(word):
    if word == None and word.text == '':
        return '없음'
    else:
        return re.sub('[\t\n]','',word.text)
    
from selenium import webdriver
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

excutable_path = "chromedriver.exe"
driver = webdriver.Chrome(executable_path=excutable_path)

base_url = 'https://ko.wikipedia.org/'

search_seit = urlmake(base_url)
search_list_url = search_seit.select_one('#n-recentchanges > a').attrs['href']

search_lists = urlmake(base_url + search_list_url).select('.mw-title')


columns = ["title", "category", "content_text"]
df = pd.DataFrame(columns=columns)

for i in search_lists:
    if i != None:
        
        list_url = urlmake(base_url + i.select_one('a').attrs['href'])
        title = list_url.select_one('h1').text
        category = "/ ".join(list(map(isNone,list_url.select('#mw-normal-catlinks > ul > li')))) or"없음"
        content_text = isNone(list_url.select_one('.mw-parser-output '))
        row = [title, category, content_text]
        series = pd.Series(row, index=df.columns)
        df = df.append(series, ignore_index=True)
        
driver.close()
```

```python
def text_cleaning(text):
    # 한글과 띄어쓰기를 제외한 모든 글자 패턴을 지정하여 hangul로 정의
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') 
    result = hangul.sub('', text) # 한글과 띄어쓰기를 제외한 모든 글자 패턴을 '' 빈 문자로 치환
    return result

df['title'] = df['title'].apply(text_cleaning)
df['category'] = df['category'].apply(text_cleaning)
df['content_text'] = df['content_text'].apply(text_cleaning)

df.head(5)
```

```python
title_corpus = ''.join(df['title'].tolist())
category_corpus = ''.join(df['category'].tolist())
content_corpus = ''.join(df['content_text'].tolist())

import konlpy

from konlpy.tag import Okt
from collections import Counter

# konlpy의 형태 분석기로 명사 단위의 키워드를 추출합니다.
nouns_tagger =Okt()
nouns = nouns_tagger.nouns(content_corpus)
count = Counter(nouns)
```

```python
korean_stopwords_path = 'korean_stopwords.txt'

# 텍스트 파일을 오픈 
with open(korean_stopwords_path,encoding = 'utf8') as f: # 불용어 텍스트 파일을 열어 f로 치환
    stopwords = f.readlines() # 파일로 불용어를 한줄씩 읽어들임
stopwords = [x.strip() for x in stopwords] # 리스트 생성
print(stopwords[:10]) #리스트 출력
```

```python
#나무위키 페이지에 맞는 불용어를 추가
namu_wiki_stopwords = ['상위','문서','내용','누설','아래','해당','설명','표기','추가','모든','사용','매우','가장',
                      '줄거리','요소','상황','편집','틀','경우','때문','모습','정도','이후','사실','생각','인물','이름','년월']
#나무위키 불용어를 기존 불용어 사전에 하나씩 추가
for stopword in namu_wiki_stopwords:
    stopwords.append(stopword)
```

```python
# 키워드 데이터에서 불용어를 제거
# Counter는 딕셔너리로 x는 키 ,remove_char_counter[x]값으로 x는 키워드,remove_char_counter[x]는 빈도수
# 기존 한글자 키워드를 제거한 content_corpus의 키워드에서 불용어 사전에 있지 않은 키워드만 remove_char_counter에 저장

remove_content = Counter({x : count[x] for x in count if len(x) > 1})
remove_content = Counter({x : remove_content[x] for x in count if x not in stopwords})

nouns_tagger =Okt()
nouns = nouns_tagger.nouns(category_corpus)
count = Counter(nouns)

remove_category = Counter({x : count[x] for x in count if len(x) > 1})
remove_category = Counter({x : remove_category[x] for x in count if x not in stopwords})

nouns_tagger =Okt()
nouns = nouns_tagger.nouns(title_corpus)
count = Counter(nouns)

remove_title = Counter({x : count[x] for x in count if len(x) > 1})
remove_title = Counter({x : remove_title[x] for x in count if x not in stopwords})

print(remove_content)
```

```python
import random
import pytagcloud
import webbrowser

# 가장 출현 빈도수가 높은 40개의 단어를 선정
ranked_tags = remove_content.most_common(40)

# pytagcloud로 출력할 40개의 단어를 입력. 단어 출력의 최대 크기는 80으로 제한
taglist = pytagcloud.make_tags(ranked_tags, maxsize=80)

pytagcloud.create_tag_image(taglist, 'wordcloud_heart.jpg', size=(900, 600), fontname='NanumGothic', rectangular=False)
```
