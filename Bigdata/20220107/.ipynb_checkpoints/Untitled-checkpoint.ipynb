# [검색어에 따라 다른 결과 나타내기]
    
    - 프로토콜,도메인,패스(경로),파타미터(서버의 추가적인 정보를 제공하기 위해서 사용, 키와 값으로 구성, &로 구분)
    - query는 검색어에 해당됨
    - 파이썬에서 사용자에게 검색어를 입력받은 뒤 query값에 적용하면 검색어 변경가능

```python
# 파이썬 기본 내장함수 input으로 사용자로 부터 검색어 입력받음
keyword = input('insert keyword, 검색어를 입력하세요')
#뉴스의 URL로부터 응답 요청
#query의 값으로 사용자로부터 입력받은 keyword가 대입되도록 함
response = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query='+keyword)
#웹 사이트의 html 소스 코드를 가져옴
html = response.text
#HTML 텍스트를 가져오기 쉽게  soup를 적용
#html.parser(html 번역기로 해석)
soup = BeautifulSoup(html,'html.parser')

#slelct는 여러개의 태그를 가져옴, select_one으로 원하는 태그 하나를 가져옴
#slelct로 뉴스 기사 10개를 다 가져옴
#soup.select_one('.news_tit')
#결과가 리스트 형태로 나옴
links = soup.select('.news_tit') 

#리스트의 값을 하나씩 for문으로 출력
for link in links:
    title = link.text # 태그 내 텍스트 요소 가져옴
    url = link.attrs['href']  #속성 href의 값을 가져옴
    print(title+'\n'+url)
```

```python
movies = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')
#웹 사이트의 html 소스 코드를 가져옴
moviesHtml = movies.text
#HTML 텍스트를 가져오기 쉽게  soup를 적용
#html.parser(html 번역기로 해석)
moviesSoup = BeautifulSoup(moviesHtml,'html.parser')

moviesLinks = moviesSoup.select('tr') 

# #리스트의 값을 하나씩 for문으로 출력
for link in moviesLinks:
    movie = link.select_one('.tit3>a')
    if movie != None:
        nums = link.select_one('.ac>img')
        num = nums.attrs['alt']
        title = movie.text
        url = movie.attrs['href']  #속성 href의 값을 가져옴
        print(f'{num}.\t{title}\n{url}')
```

# [pyautogui 라이브러리로 간단한 팝업창 띄우기]

- 마우스, 키보드 매크로 라이브러리
- 간단한 팝업창 띄우기 기능
- pyautogui.prompt('검색어 입력')

```python
import pyautogui

keyword = pyautogui.prompt('검색어 입력')
#뉴스의 URL로부터 응답 요청
#query의 값으로 사용자로부터 입력받은 keyword가 대입되도록 함
response = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query='+keyword)
#웹 사이트의 html 소스 코드를 가져옴
html = response.text
#HTML 텍스트를 가져오기 쉽게  soup를 적용
#html.parser(html 번역기로 해석)
soup = BeautifulSoup(html,'html.parser')

#slelct는 여러개의 태그를 가져옴, select_one으로 원하는 태그 하나를 가져옴
#slelct로 뉴스 기사 10개를 다 가져옴
#soup.select_one('.news_tit')
#결과가 리스트 형태로 나옴
links = soup.select('.news_tit') 

#리스트의 값을 하나씩 for문으로 출력
for link in links:
    title = link.text # 태그 내 텍스트 요소 가져옴
    url = link.attrs['href']  #속성 href의 값을 가져옴
    print(title+'\n'+url)
```

# [실전 프로젝트] 뉴스 데이터 수집하기/
# 0.3여러 페이지 결과 모두 가져오기


- URL 분석하기, 여러 페이지의 결과 가져오기 위해 이동하면서 URL이 어떻게 변하는지 살펴보기
- 뉴스 다음 페이지로 이동시 start 파라미터의 값이 10씩 증가
- 1페이지: start = 1, 2페이지 start = 11, 3페이지 start = 21....

```python
# 파이썬 기본 내장함수 input으로 사용자로부터 검색어 입력받음
# keyword = input("insert keyword, 검색어를 입력하세요.")

# pyautogui로 팝업창 띄워 사용자로부터 검색어 입력받기
keyword = pyautogui.prompt("검색어를 입력하세요.")
listNum = 1
pageNum = 1

for i in range(1, 30, 10):
    # URL로부터 응답 요청
    # query의 값으로 사용자로부터 입력받은 keyword가 대입되도록 함
    # f-string 사용, f"문자열{변수}"
    response = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={keyword}&start={i}")
    # 웹 사이트의 html 소스 코드를 가져옴
    html = response.text
    # HTML 텍스트를 가져오기 쉽게 soup를 적용
    # html.parser (html 번역기로 해석)
    soup = BeautifulSoup(html, 'html.parser')

    # select는 여러개의 태그를 가져옴, select_one으로 원하는 태그 하나만 가져옴
    # select로 뉴스 기사 10개를 다 가져옴
    # soup.select_one(".news_tit")
    links = soup.select(".news_tit") # 결과가 리스트 형태로 나옴

    # 리스트의 값을 하나씩 for 문으로 출력
    for link in links:
        title = link.text # 태그 내 텍스트 요소 가져옴
        url = link.attrs['href'] # 속성 href 의 값을 가져옴
        print(f'{i}. {title}{url}')
    print(f"------------------------------------------------ {pageNum} 페이지 ------------------------------------------------------- ")
    pageNum += 1
```
