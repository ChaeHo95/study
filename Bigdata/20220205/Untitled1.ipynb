```python
# paras로 서비스키를 입력 할 경우 Decoding 된 key를 입력 할 것
url = 'http://apis.data.go.kr/B552584/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty'

param_data = {
    'sidoName':'대구',
    'pageNo':'1',
    'numOfRows':'1000',
    'returnType':'json',
    'serviceKey':'nQ3X8tkoYMoAKW5srgV9gbOgaQBGt/wyJ3g/p/yqK8E+V5DYtPliLmupeGydOGaJsSnF6lyRaCyCD8/Cc6rDrg==',
    'var':'1.0'}

res = requests.get(url,params=param_data)
```

```python
# serviceKey가 Encoding 된 것을 볼 수 있음
print(res.url)
```

```python
# json형태로 보내준 걸 알 수 있음
print(res.text)
```

```python
# .json => json 형태를 dictionary로 가져옴
data = res.json()
```
