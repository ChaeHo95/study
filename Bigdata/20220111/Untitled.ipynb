```python
import requests
import numpy as np
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup
resonse = requests.get('https://movie.naver.com/movie/running/current.naver')
html = resonse.text
soup = BeautifulSoup(html,'html.parser')

movies = soup.select('.lst_dsc') 

num = 0

xs = []
ys = []

for movie in movies:
    movie_name = movie.select_one('.tit>a').text
    if(movie_name != None):
        scope = movie.select_one('.num').text
        print(f'{num+1}. {movie_name} / 별점: {scope}')
        num += 1
        xs.append(movie_name)
        ys.append((float)(scope))
        if(num == 10):
            break

print()

plt.figure(figsize=(10,6))
plt.title('영화 별점그래프')
plt.xlabel('영화 이름')
plt.ylabel('별점')
plt.xticks(rotation=70)

# 하나의 배열로 만들어 하나씩 비교 할 수 있게 함
plts = plt.bar(xs, ys, width=0.6, color='skyblue')

print(xs)
print(ys)

ysMax = max(ys)

for pl in plts:
     if(pl.get_height() >= ysMax):
            pl.set_color("pink")

plts
```
